# Data Visualization using pandas (multiple plots)

a simple code to plot multiple graphs which are linked to each other

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
averagemarks = [15, 25, 30]
class10 = [12, 26, 38]
class9 = [23, 37, 40]
plt.plot(averagemarks, class10 )
plt.plot(averagemarks, class9 )
plt.title("plot")
plt.xlabel("average marks")
plt.ylabel("class 9 & class 10")
plt.legend(["class 10", "class 9"])
plt.show
```
