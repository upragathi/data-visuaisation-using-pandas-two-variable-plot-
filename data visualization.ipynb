# Data Visualisation using Pandas

example 1

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x=[2, 3, 5]
y=[1, 5, 7]
plt.plot(x, y)
plt.title("average marks")
plt.xlabel("average number of students")
plt.ylabel("marks")
plt.show
```

example 2

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
voltage=[-85, -80, -20, 0, 0.2, 0.4, 0.6, 0.7, 0.8]
current=[-30, -10, -8, 0, 0.2, 0.4, 0.6, 10, 100]
plt.plot(voltage, current)
plt.title("sample VI characteristics of a diode")
plt.xlabel("voltage (in volts)")
plt.ylabel("current (in milli amps)")
plt.show
```
