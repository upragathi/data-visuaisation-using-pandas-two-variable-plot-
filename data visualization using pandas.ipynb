# DATA VISUALIZATION USING PANDAS

A basic example on visualizing data (here plotting an X-Y graph) using <font color='green'>Pandas </font>

example 1

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
X = [1.5, 3.7, 6.0]
Y = [3.4, 6.0, 8.9]
plt.plot(X, Y)
plt.show
```
